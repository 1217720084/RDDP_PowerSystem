```python
import pandas as pd
from seaborn import *
from statsmodels import *
import matplotlib.pyplot as plt
```

## Storage approximation analysis

```python
df = pd.read_csv("../results/ess_approximation.csv")
df = df[df["gap"]>=0]
hydro = df[df["efficiency"]<0.85]
battery = df[(df["efficiency"]>=0.85)&(df["efficiency"]<0.95)]
ideal = df[df["efficiency"]>=0.95]
```

```python
set(rc={'figure.figsize':(12.7,9.27)},font_scale=2)
set_style("ticks")
distplot(hydro["gap"],hist=False,kde_kws=
         {"color": "r", "lw": 1, "label": "hydro-unit\n$\mu=0.021$\n$\sigma=0.010$","shade":True,"cut":0})
distplot(battery["gap"],hist=False,kde_kws=
         {"color": "g", "lw": 1, "label": "battery storage\n$\mu=0.016$\n$\sigma=0.008$","shade":True,"cut":0})
distplot(ideal["gap"],hist=False,
         kde_kws={"color": "b", "lw": 1, "label": "ideal storage\n$\mu=0.007$\n$\sigma=0.007$","shade":True,"cut":0})
plt.savefig("../fig/M_sigma.pgf")
```

```python
set(rc={'figure.figsize':(10.7,7.27)},font_scale=2)
set_style("ticks")
distplot(hydro["p"],hist=False,kde_kws=
         {"color": "r", "lw": 1, "label": "hydro-unit\n$\mu=0.027,\sigma=0.047$","shade":True,"cut":0})
distplot(battery["p"],hist=False,kde_kws=
         {"color": "g", "lw": 1, "label": "battery storage\n$\mu=0.002,\sigma=0.007$","shade":True,"cut":0})
distplot(ideal["p"],hist=False,
         kde_kws={"color": "b", "lw": 1, "label": "ideal storage\n$\mu=0.002,\sigma=0.008$","shade":True,"cut":0})
plt.savefig("M_p.pgf")
```
